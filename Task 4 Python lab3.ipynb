```python
def task10(N: int):
  for a in range(1, int(N**0.5) + 1):
    b2 = N - a*a
    b = int(b2**0.5)
    if b > 0 and b*b == b2:
      print(f"{a}^2 + {b}^2 = {N}")
      return

  print(f"Число {N} не можна представити квадратами цілих чисел")

while True:
  try:
    inp = input("Введіть натуральне число N: ")
    N = int(inp)

    if N <= 1:
      print("Натуральне число не може бути менше 1")
      continue
    break

  except ValueError:
    print("Помилка. Присутнє нечислове значення, або дробове число")


task10(N)
```
